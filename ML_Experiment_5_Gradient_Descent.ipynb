```python
#12 Feb
import numpy as np
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
```

```python
X = np.array([5,7,12,16,20])
Y = np.array([40,120,180,210,240])
m = 0
c = 0
n = 5
alpha = 0.1
```

```python
def gradient_descent(X, Y, m, c, n, alpha):
    y_pred = m * X + c
    D_m = (-2/n) * np.sum((Y - y_pred) * X)
    D_c = (-2/n) * np.sum(Y - y_pred)
    m = m - (alpha * D_m)
    c = c - (alpha * D_c)
    return m, c, D_m
e_m, e_c, D_m = gradient_descent(X,Y,m,c,n,alpha)
print(f"Final slope (m): {e_m}")
print(f"Final intercept (c): {e_c}")
print(f"Final DM: {D_m}")
```

```python
def gradient_descent_i(X, Y, m, c, n, alpha, iterations):
    for i in range(iterations):
        Y_pred = m * X + c
        D_m = (-2/n) * sum(X * (Y - Y_pred))
        D_c = (-2/n) * sum(Y - Y_pred)
        m = m - alpha * D_m  #
        c = c - alpha * D_c
    return m, c, Y_pred


iterations = 5
final_m, final_c, final_Y_pred = gradient_descent_i(X, Y, m, c, n, alpha, iterations)

print(f"Final slope (m): {final_m}")
print(f"Final intercept (c): {final_c}")
print(f"Final Y_pred (Y): {final_Y_pred}")
```

```python
data = pd.DataFrame({
    'carrot': [0.23, 0.23, 0.29, None, 0.31, 0.24, 0.24, 0.26, None],
    'cut': [None, 'premium', 'good', 'premium', 'good', 'Very good', 'Very good', 'Very good', 'good'],
    'color': ['E', 'E', 'E', 'J', 'J', 'J', 'H', 'H', 'E'],
    'price': [326, 326, 327, 334, 335, 336, 336, 337, 338]
})
data
```

```python
mean_carrot_value = np.mean(data['carrot'])
data['carrot'].fillna(mean_carrot_value, inplace=True)
data
```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

one_hot_encoder = OneHotEncoder()
encoded_features = one_hot_encoder.fit_transform(data[['cut', 'color']])
encoded_df = pd.DataFrame(
    encoded_features.toarray(),
    columns=one_hot_encoder.get_feature_names_out(['cut', 'color'])
)
final_data = pd.concat([data.drop(['cut', 'color'], axis=1), encoded_df], axis=1)
print(final_data)
```

```python
import pandas as pd
dummy_encoded_df = pd.get_dummies(data[['cut', 'color']], drop_first=True)
final_data_with_dummies = pd.concat([data.drop(['cut', 'color'], axis=1), dummy_encoded_df], axis=1)
print(final_data_with_dummies)
```
