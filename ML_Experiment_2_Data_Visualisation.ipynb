## Outlier detection


```python
def outlier_detection(column):
  Q1 = column.quantile(0.25)
  Q3 = column.quantile(0.75)
  IQR = Q3-Q1
  lower_fence = Q1-1.5*IQR
  upper_fence = Q3 + 1.5*IQR
  outliers = (column<lower_fence) | (column>upper_fence)
  return outliers
```

```python
outliers_iqr = dataframe.apply(outlier_detection)
print("Outliers: \n",outliers_iqr)

threshold = 3
zoutliers = (abs(zscore)>threshold)
print("ZScore outliers: ",zoutliers)
```
